```python
import pandas as pd
import numpy as np
df1=pd.read_csv('/content/Train82.csv.zip')
df2=pd.read_csv('/content/Test82 2.csv.zip')
```

```python
df1.dropna(inplace = True)
df1.reset_index(drop=True, inplace=True)
df1
```

```python
df2.dropna(inplace = True)
df2.reset_index(drop=True, inplace=True)
df2
```

```python
X = df1.drop(columns = ['label'])
y = df1['label']
```

```python
y.unique().shape
```

```python
X=X.drop(columns = ['ImgPath'])
```

```python
x_temp=X
y_temp=y
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
scaler = StandardScaler()
X_std = scaler.fit_transform(x_temp)
```

```python
import keras
from keras import backend as K
from keras.models import Sequential
from keras.layers import Activation
from keras.layers import Dense
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
model=Sequential([
    Dense(200, input_shape=(62,), activation='relu'),
    Dense(200, activation='relu'),

    Dense(200, activation='relu'),
    Dense(82, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics= ['accuracy'])
```

```python
model.fit(X_std, y_temp, epochs=500, batch_size=100, shuffle=True, verbose=2)
```

```python
X2 = df2.drop(columns = ['ImgPath'])
y2 = df2['ImgPath']
```

```python
std_encoder = StandardScaler()
X2_std = std_encoder.fit_transform(X2)
```

```python
y_pred = model.predict(X2_std, batch_size=10, verbose=0)
```
